```python
import pandas as pd

# Create DataFrame
df = pd.DataFrame({
    "Student_ID": [101,102,103,104,105,106,107,108,109,110,
                   111,112,113,114,115,116,117,118,119,120,
                   121,122,123,124,125],

    "Name": ["Amit","Priya","Rahul","Sneha","Karan",
             "Neha","Rohit","Pooja","Vikas","Anjali",
             "Arjun","Meera","Sanjay","Riya","Deepak",
             "Komal","Nitin","Simran","Manish","Kavita",
             "Yash","Isha","Varun","Aarti","Mohit"],

    "Age": [20,21,22,20,23,
            21,22,20,24,23,
            21,22,20,23,24,
            21,22,20,23,24,
            21,22,23,20,21],

    "Course": ["B.Tech","BCA","MCA","B.Sc","B.Tech",
               "MBA","BCA","MCA","B.Tech","MBA",
               "B.Sc","BCA","MCA","MBA","B.Tech",
               "BCA","B.Sc","MBA","MCA","B.Tech",
               "BCA","MBA","B.Sc","MCA","B.Tech"],

    "Marks": [85,78,92,67,88,
              74,81,90,69,95,
              72,84,91,79,87,
              76,89,83,94,68,
              80,77,86,93,82],

    "City": ["Delhi","Mumbai","Lucknow","Jaipur","Pune",
             "Delhi","Noida","Lucknow","Indore","Bhopal",
             "Kanpur","Agra","Patna","Ranchi","Surat",
             "Chennai","Hyderabad","Kolkata","Nagpur","Delhi",
             "Jaipur","Pune","Noida","Lucknow","Mumbai"]
})

# Display DataFrame
print(df)
```

```python
import matplotlib.pyplot as plt
#Create a line chart of Student vs Marks
plt.figure(figsize=(10,5))
plt.plot(df['Name'], df['Marks'], marker='o', color='b')
plt.title('Student vs Marks')
plt.xlabel('Student Name')
plt.ylabel('Marks')
plt.xticks(rotation=50)
plt.grid(False)
```

```python
# A line chart of age and marks
plt.figure(figsize=(10,8))
plt.plot(df['Age'], df['Marks'], marker='o', color='g')
plt.title('Age vs Marks')
plt.xlabel('Age')
plt.ylabel('Marks')
plt.grid(True)

```

```python
## bar chart of average marks of each course

plt.figure(figsize=(10,5))
avg_marks = df.groupby('Course')['Marks'].mean()
avg_marks.plot(kind='pie', color='orange')
plt.title('Average Marks by Course')
plt.xlabel('Course')
plt.ylabel('Average Marks')
```

```python
plt.figure(figsize=(10,5))
city_counts = df['City'].value_counts()
city_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Students in Each City')
plt.xlabel('City')
plt.ylabel('Number of Students')
plt.grid(axis='y', linestyle='-', alpha=0.6)
#
```

```python
#Horizantal bar of top 5 student
top_student = df.nlargest(5,"Marks")
plt.plot(figsize = (5,10),facecolor='lightred')
plt.xlabel("Top student")
plt.ylabel("Marks")
plt.title("Toppers")
plt.barh(top_student["Name"], top_student["Marks"], color='orange')
```

```python
# Historgram of distrubtion of marks

plt.figure(figsize=(10,5))
plt.hist(df['Marks'], bins=10, color='purple', edgecolor='black')
plt.title('Distribution of Marks')
plt.xlabel('Marks')
plt.ylabel('Number of Students')
```

```python
## Histogram of Age
plt.figure(figsize=(10,6))
plt.hist(df["Age"], bins = 6 , color = 'yellow', edgecolor = 'black')
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Histogram")
```

```python
plt.scatter(df["Age"] , df["Marks"] , color = 'black')
plt.title("Age vs Marks")
plt.xlabel("age")
plt.ylabel("Marks")
```

```python
plt.figure(figsize=(10,5))
plt.scatter(
    df["Age"],
    df["Marks"],
    s = df["Marks"],
    c = df["Age"],
    edgecolor = "black"
)
plt.title("Age vs Marks")
plt.xlabel("Age")
plt.ylabel("avg_marks")
plt.colorbar(label='Age')



```

```python
import numpy as np

# Q10: Percentage of students in each Course
plt.figure(figsize=(6,6))
df['Course'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Percentage of Students by Course')

# Q11: Pie chart of students from each City with explode/shadow/startangle/autopct
plt.figure(figsize=(8,8))
explode = [0.03]*len(city_counts)
plt.pie(city_counts.values, labels=city_counts.index, explode=explode, shadow=True, startangle=90, autopct='%1.1f%%')
plt.title('Students by City')

# Q12: Box plot for Marks
plt.figure(figsize=(6,6))
plt.boxplot(df['Marks'], vert=True, patch_artist=True)
plt.title('Box Plot of Marks')
plt.ylabel('Marks')

# Q13: Box plots of Marks by Course
courses = avg_marks.index.tolist()
data_by_course = [df[df['Course']==c]['Marks'] for c in courses]
plt.figure(figsize=(8,6))
bp = plt.boxplot(data_by_course, tick_labels=courses, patch_artist=True, notch=True, showmeans=True)
for patch, color in zip(bp['boxes'], plt.cm.tab20.colors):
    patch.set_facecolor(color)
plt.title('Marks by Course')
plt.ylabel('Marks')

# Q14: Area chart of Marks sorted by Student_ID
sorted_df = df.sort_values('Student_ID')
plt.figure(figsize=(10,4))
plt.fill_between(sorted_df['Student_ID'], sorted_df['Marks'], alpha=0.4)
plt.plot(sorted_df['Student_ID'], sorted_df['Marks'], marker='o')
plt.title('Marks (Area) sorted by Student_ID')
plt.xlabel('Student_ID')
plt.ylabel('Marks')

# Q15: Stack plot of cumulative marks for three Courses (choose first 3 from avg_marks)
pivot_course = df.pivot_table(index='Student_ID', columns='Course', values='Marks', aggfunc='sum').fillna(0)
chosen_courses = avg_marks.index[:3].tolist()
x = pivot_course.index
ys = [pivot_course[c] if c in pivot_course.columns else np.zeros_like(x) for c in chosen_courses]
plt.figure(figsize=(10,5))
plt.stackplot(x, *ys, labels=chosen_courses)
plt.legend(title='Course')
plt.title('Stack Plot: Marks by Course (3 courses)')
plt.xlabel('Student_ID')

# Q16: Line chart average Marks for each Age (sorted)
age_avg = df.groupby('Age')['Marks'].mean().sort_index()
plt.figure(figsize=(8,4))
plt.plot(age_avg.index, age_avg.values, marker='o')
plt.title('Average Marks by Age')
plt.xlabel('Age')
plt.ylabel('Average Marks')

# Q17: Two lines: Average Marks and Maximum Marks by Course
max_marks = df.groupby('Course')['Marks'].max().reindex(courses)
plt.figure(figsize=(8,5))
plt.plot(courses, avg_marks.values, marker='o', label='Average', linestyle='-')
plt.plot(courses, max_marks.values, marker='s', label='Maximum', linestyle='--')
plt.title('Average and Maximum Marks by Course')
plt.xlabel('Course')
plt.ylabel('Marks')
plt.legend()

# Q18: Grouped bar chart: Average vs Maximum Marks per Course
x = np.arange(len(courses))
width = 0.35
plt.figure(figsize=(10,5))
plt.bar(x - width/2, avg_marks.values, width, label='Average')
plt.bar(x + width/2, max_marks.values, width, label='Maximum')
plt.xticks(x, courses)
plt.ylabel('Marks')
plt.title('Avg vs Max Marks by Course')
plt.legend()

# Q19: Bar chart Top 10 students by Marks (green if >=85 else red)
top10 = df.nlargest(10, 'Marks').sort_values('Marks')
colors = ['green' if m >= 85 else 'red' for m in top10['Marks']]
plt.figure(figsize=(10,5))
plt.bar(top10['Name'], top10['Marks'], color=colors)
plt.xticks(rotation=45)
plt.title('Top 10 Students by Marks')

# Q20: Horizontal bar chart of cities sorted by average Marks
city_avg = df.groupby('City')['Marks'].mean().sort_values()
plt.figure(figsize=(8,6))
plt.barh(city_avg.index, city_avg.values, color='skyblue')
plt.title('Average Marks by City')
plt.xlabel('Average Marks')

# Q21: Two overlapping histograms: Marks and Age
plt.figure(figsize=(8,5))
plt.hist(df['Marks'], bins=10, alpha=0.6, label='Marks', color='blue')
plt.hist(df['Age'], bins=6, alpha=0.5, label='Age', color='orange')
plt.legend()
plt.title('Overlapping Histograms: Marks and Age')

# Q22: Histogram of Marks with density=True, cumulative=True, 8 bins
plt.figure(figsize=(8,4))
plt.hist(df['Marks'], bins=8, density=True, cumulative=True, color='green', alpha=0.6, edgecolor='black')
plt.title('Cumulative Density Histogram of Marks (8 bins)')
plt.xlabel('Marks')
plt.ylabel('Cumulative Density')
print("Q22: The plot shows the cumulative density (proportion) of students with Marks ≤ bin edge.")

# Q23: Scatter plot X=Age, Y=Marks, Size=Marks*5, Color=Age, Colormap='viridis'
plt.figure(figsize=(8,6))
sc = plt.scatter(df['Age'], df['Marks'], s=df['Marks']*5, c=df['Age'], cmap='viridis', edgecolor='k')
plt.colorbar(sc, label='Age')
plt.title('Age vs Marks (size ∝ Marks)')
plt.xlabel('Age')
plt.ylabel('Marks')

# Q24: Highlight students scoring above 85 in red, others in blue
plt.figure(figsize=(8,6))
high = df['Marks'] > 85
plt.scatter(df.loc[~high, 'Age'], df.loc[~high, 'Marks'], color='blue', label='<=85')
plt.scatter(df.loc[high, 'Age'], df.loc[high, 'Marks'], color='red', label='>85')
plt.legend()
plt.title('Highlight Scores > 85')
plt.xlabel('Age')
plt.ylabel('Marks')

# Q25: Doughnut chart of students in each Course
plt.figure(figsize=(6,6))
sizes = df['Course'].value_counts()
wedges, texts, autotexts = plt.pie(sizes, labels=sizes.index, autopct='%1.1f%%', startangle=90)
plt.gca().add_artist(plt.Circle((0,0), 0.40, color='white'))
plt.title('Doughnut Chart: Students by Course')

# Q26: Box plots of Marks for every City (show mean, notch, different colors)
cities = city_counts.index.tolist()
data_by_city = [df[df['City']==c]['Marks'] for c in cities]
plt.figure(figsize=(12,6))
bp = plt.boxplot(data_by_city, patch_artist=True, notch=True, showmeans=True)
plt.xticks(range(1, len(cities)+1), cities, rotation=45)
colors = plt.cm.tab20.colors
for i, box in enumerate(bp['boxes']):
    box.set_facecolor(colors[i % len(colors)])
plt.xticks(rotation=45)
plt.title('Marks by City')

# Q27: Area chart sorted by Student_ID filled only where Marks > 80
sorted_df = df.sort_values('Student_ID')
x = sorted_df['Student_ID']
y = sorted_df['Marks']
plt.figure(figsize=(10,4))
plt.plot(x, y, marker='o')
plt.fill_between(x, y, where=(y>80), interpolate=True, color='orange', alpha=0.5)
plt.title('Area fill where Marks > 80')
plt.xlabel('Student_ID')
plt.ylabel('Marks')

# Q28: Stack plot cumulative Marks of students from three Cities (choose first 3 cities)
pivot_city = df.pivot_table(index='Student_ID', columns='City', values='Marks', aggfunc='sum').fillna(0)
chosen_cities = city_counts.index[:3].tolist()
x = pivot_city.index
ys_c = [pivot_city[c] if c in pivot_city.columns else np.zeros_like(x) for c in chosen_cities]
plt.figure(figsize=(10,5))
plt.stackplot(x, *ys_c, labels=chosen_cities)
plt.legend(title='City')
plt.title('Stack Plot: Marks by City (3 cities)')
plt.xlabel('Student_ID')

# Q29: Heatmap of correlation between Student_ID, Age, and Marks
corr = df[['Student_ID','Age','Marks']].corr()
plt.figure(figsize=(6,5))
plt.imshow(corr, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.index)), corr.index)
plt.title('Correlation Heatmap: Student_ID, Age, Marks')

plt.tight_layout()
```
