##### Questions to solve

1. Create an array to store prices of 15 products.
2. Input all product prices.
3. Display all product prices.
4. Calculate the total cost of all products.
5. Find the average product price.
6. Find the most expensive product.
7. Find the cheapest product.
8. Count products costing more than ₹1000.
9. Count products costing less than ₹500.
10. Search for a particular product price.
11. Sort prices in ascending order.
12. Sort prices in descending order.
13. Apply a 20% discount to all prices.
14. Replace a product price with a new value.
15. Display the final updated price list.

```python
# Question 1: Crete an array to store prices of 15 product.
import numpy as np
prices = np.zeros(15, dtype=float)

## Inputting elements, as it is second question
for i in range(0 , 14):
    prices[i] = float(input(f"Enter price of product {i+1}: "))

## Print all the prices 
print("Prices of products:", prices)

## Average of product 
avg_price = np.mean(prices)
print("Average price of products:", avg_price)

## Most expeinve 
most_expensive = np.max(prices)
print("Most expensive product price:", most_expensive)

## most cheap
most_cheap = np.min(prices)
print("Most cheap product price:", most_cheap)

## product cositng more than 1000
expensive_products = prices[prices > 1000]
print("Products costing more than 1000:", expensive_products)

## Less than 500 product
less_than500 = prices[prices < 500]
print("Products costing less than 500 :", less_than500)

## Search for particular 

search = int(input("Enter the price you want to search for: "))
if search in prices:
    print(f"Product with price {search} found.")
else:
    print(f"Product with price {search} not found.")


## Sort in ascending

ascending = np.sort(prices)
print("Prices sorted in ascending order:", ascending)

## Sort in Descending 

descending = np.sort(prices)[::-1]
print("Prices sorted in descending order:", descending)

## disount of 20% 
print("Prices after 20% discount:", prices * 0.8)

## replace a proudct pirce with new price
replace_index = int(input("Enter the index of the product you want to replace (0-14): "))
new_price = float(input("Enter the new price: "))
if 0 <= replace_index < 15:
    prices[replace_index] = new_price
    print(f"Product price at index {replace_index} replaced with {new_price}.")
else:   
    print("Invalid index. Please enter a value between 0 and 14.")

print("Updated prices of products:", prices)


```
