```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Student_ID": [101,102,103,104,105,106,107,108,109,110,
                   111,112,113,114,115,116,117,118,119,120],

    "Name": ["Aman","Riya",None,"Karan","Priya","Rahul","Neha",None,
             "Vikas","Pooja","Ankit","Simran","Rohit",None,"Kajal",
             "Deepak","Meena","Arjun","Nisha",None],

    "Age": [20,21,np.nan,22,19,23,21,20,np.nan,22,
            24,23,21,20,22,np.nan,19,21,23,20],

    "City": ["Delhi","Mumbai","Pune",None,"Jaipur","Delhi","Mumbai",
             "Pune","Delhi",None,"Jaipur","Mumbai","Delhi","Pune",
             None,"Delhi","Mumbai","Jaipur","Pune","Delhi"],

    "Marks": [85,90,78,np.nan,88,76,91,82,95,np.nan,
              80,84,79,92,87,75,np.nan,89,81,93]
})

print(df)
```

| Level | No. | Condition |
|---|---:|---|
| Basic | 1 | Show all students whose Marks > 85 |
| Basic | 2 | Show all students whose Age is 20 |
| Basic | 3 | Show all students who live in Delhi |
| Basic | 4 | Show all students who live in Mumbai |
| Basic | 5 | Show all students whose Salary > 35000 |
| Basic | 6 | Show all students whose Marks < 80 |
| Basic | 7 | Show all students whose Age > 21 |
| Basic | 8 | Show all students whose Department is CS |
| Basic | 9 | Show all students whose Department is IT |
| Basic | 10 | Show all students whose City is Lucknow |
| Medium | 11 | Show students whose Marks > 80 AND Age < 21 |
| Medium | 12 | Show students from Delhi AND Marks > 85 |
| Medium | 13 | Show students from Mumbai OR Pune |
| Medium | 14 | Show students NOT from Lucknow |
| Medium | 15 | Show students whose Marks between 80 and 90 |
| Medium | 16 | Show students whose Department is CS OR EC |
| Medium | 17 | Show students whose Salary > 32000 AND Department is IT |
| Medium | 18 | Show students whose Age is 21 OR Marks > 90 |
| Medium | 19 | Show students whose City is Delhi, Mumbai, or Lucknow |
| Medium | 20 | Show students whose Marks > 85 AND Age >= 20 |

```python
print("1. Marks > 85:\n", df[df["Marks"] > 85])
print("2. Age == 20:\n", df[df["Age"] == 20])
print("3. City == Delhi:\n", df[df["City"] == "Delhi"])
print("4. City == Mumbai:\n", df[df["City"] == "Mumbai"])
print("5. Salary > 35000:\n", df[df["Salary"] > 35000])
print("6. Marks < 80:\n", df[df["Marks"] < 80])
print("7. Age > 21:\n", df[df["Age"] > 21])
print("8. Department == CS:\n", df[df["Department"] == "CS"])
print("9. Department == IT:\n", df[df["Department"] == "IT"])
print("10. City == Lucknow:\n", df[df["City"] == "Lucknow"])

print("11. Marks > 80 and Age < 21:\n", df[(df["Marks"] > 80) & (df["Age"] < 21)])
print("12. City == Delhi and Marks > 85:\n", df[(df["City"] == "Delhi") & (df["Marks"] > 85)])
print("13. City == Mumbai or Pune:\n", df[df["City"].isin(["Mumbai", "Pune"])])
print("14. Not from Lucknow:\n", df[df["City"] != "Lucknow"])
print("15. Marks between 80 and 90:\n", df[df["Marks"].between(80, 90)])
print("16. Department == CS or EC:\n", df[df["Department"].isin(["CS", "EC"])])
print("17. Salary > 32000 and Department == IT:\n", df[(df["Salary"] > 32000) & (df["Department"] == "IT")])
print("18. Age == 21 or Marks > 90:\n", df[(df["Age"] == 21) | (df["Marks"] > 90)])
print("19. City in Delhi, Mumbai, Lucknow:\n", df[df["City"].isin(["Delhi", "Mumbai", "Lucknow"])])
print("20. Marks > 85 and Age >= 20:\n", df[(df["Marks"] > 85) & (df["Age"] >= 20)])
```
