```python
# String slicing syntax:
# string[start:end:step]
# - start: index where slicing begins (inclusive)
# - end: index where slicing stops (exclusive)
# - step: how many characters to skip each time

print("Original string:", repr(s))

print("s[1:4]  ->", s[1:4])   # characters from index 1 to 3
print("s[:4]   ->", s[:4])    # from the beginning to index 3
print("s[4:]   ->", s[4:])    # from index 4 to the end
print("s[::2]  ->", s[::2])   # every second character
print("s[::-1] ->", s[::-1])  # reverse the string
print(s[3:7])  # characters from index 3 to 6
print(s[-5:-2:1])
```
