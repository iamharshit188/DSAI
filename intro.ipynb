```python
# String slicing syntax:
# string[start:end:step]
# - start: index where slicing begins (inclusive)
# - end: index where slicing stops (exclusive)
# - step: how many characters to skip each time

print("Original string:", repr(s))

print("s[1:4]  ->", s[1:4])   # characters from index 1 to 3
print("s[:4]   ->", s[:4])    # from the beginning to index 3
print("s[4:]   ->", s[4:])    # from index 4 to the end
print("s[::2]  ->", s[::2])   # every second character
print("s[::-1] ->", s[::-1])  # reverse the string
print(s[3:7])  # characters from index 3 to 6
print(s[-5:-2:1])
```

### Swtich case program to calculate Factoiral , Sum MaX Min Avg

```python
def CalculateFactorial(n):
    if n == 0:
        return 1
    else : 
        return n * CalculateFactorial(n-1)
def CalculateSumOfNumbers (n):
    if n == 0:
        return 0
    else :
        return n + CalculateSumOfNumbers(n-1)

def FindMax(n1 , n2):
    if n1 > n2:
        return n1
    else :
        return n2

def FindMin(n1 , n2):
    if n1 < n2:
        return n1
    else :
        return n2
def FindAvg(n1, n2 , n3):
    return (n1+n2+n3)/3
    
action = int(input("Enter 1 for factorial or 2 for sum: "))
if action == 1:
    n = int(input("Enter a number to calculate fact"))
    print("Factorial of", n, "is", CalculateFactorial(n))
elif action == 2:
    n = int(input("Enter a number to calculate  sum"))
    print("Sum of numbers up to", n, "is", CalculateSumOfNumbers(n))
elif action == 3:
    n1 = int(input("Enter first number to calculate Maximum Number"))
    n2 = int(input("Enter second number to calculate Maximum Number"))
    print(f"Maximum number among {n1} and {n2} is {FindMax(n1,n2)}")

else : 
    print("Invalid action selected.")

```

```python
courses = "Data Science "

class Student:
    def __init__(self, rn, n, c, p, ds):
        self.rollNo = rn
        self.name = n
        self.CLang = c
        self.python = p
        self.dataScience = ds

    def displayStudentReport(self):
        tot = self.CLang + self.python + self.dataScience
        avg = tot / 3
        per = (tot / 300) * 100

        print("\n" + "=" * 40)
        print(" " * 12 + "STUDENT REPORT CARD")
        print("=" * 40)
        print(f"Roll No     : {self.rollNo}")
        print(f"Name        : {self.name}")
        print(f"C Language  : {self.CLang}")
        print(f"Python      : {self.python}")
        print(f"{courses.strip():12}: {self.dataScience}")
        print("-" * 40)
        print(f"Total       : {tot}")
        print(f"Average     : {avg:.2f}")
        print(f"Percentage  : {per:.2f}%")

        if per >= 80:
            grade = "A"
        elif per >= 60:
            grade = "B"
        elif per >= 40:
            grade = "C"
        else:
            grade = "D"

        print(f"Grade       : {grade}")
        print("=" * 40 + "\n")


s1 = Student(1, "Alice", 85, 90, 88)
s2 = Student(2, "Bob", 72, 68, 75)

s1.displayStudentReport()
s2.displayStudentReport()
```

```python
## Method overloading in python..

# Python doesn't support traditional method overloading like Java or C++.
# Instead, we can use default arguments or *args/**kwargs to achieve similar functionality.

class Calculator:
    # Using default arguments
    def add(self, a, b=0, c=0):
        return a + b + c

# Example 1: Adding two numbers
calc = Calculator()
print("Add two numbers:", calc.add(5, 10))        # Output: 15

# Example 2: Adding three numbers
print("Add three numbers:", calc.add(5, 10, 15))  # Output: 30

# Using *args for variable number of arguments
class Printer:
    def display(self, *args):
        for item in args:
            print(item, end=" ")
        print()

# Example 3: Variable number of arguments
printer = Printer()
printer.display("Hello")                           # Output: Hello
printer.display("Hello", "World", "Python")        # Output: Hello World Python
```
